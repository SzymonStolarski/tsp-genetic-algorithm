```python
from components.data_reader import DataReader
cost_data = DataReader.read_txt(path='data/berlin52.txt', dlm=' ', data_type=int)
print(len(cost_data))
```

```python
from components.tsp_algorithms.genetic_algorithm import GeneticAlgorithmTsp
from components.ga_components.selection import RouletteSelector, TournamentSelector, ElitismSelector
from components.ga_components.crossover import PMXCrossover
from components.ga_components.mutation import SwapMutator, InverseMutator
```

```python
genetic_algorithm = GeneticAlgorithmTsp(population_size=100, n_iterations=7000000,
selector=TournamentSelector(k_percent=0.1), crossover=PMXCrossover(0.75), mutator=InverseMutator(0.05),
atomic_bomb=3000)

genetic_algorithm.learn(cost_data)
```

```python
import matplotlib.pyplot as plt
plt.plot(genetic_algorithm.results.keys(), genetic_algorithm.results.values())
plt.show()
```
